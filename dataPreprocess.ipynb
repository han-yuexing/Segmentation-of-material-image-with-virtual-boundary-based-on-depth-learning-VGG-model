```python
##you sui ji qu kuai

import cv2
from PIL import Image
import numpy as np

palette = {(0,0,0) : 0,
           (128,0,0) : 1,  ##blue:tiaowen
           (0,128,0) : 2,   ##green:kuai
           (0,0,128) : 255,
           (128,0,128) : 3,
           (0,128,128) : 4,
           (128,128,128) : 5,
           (255,255,255) : 6 ##white:padding
          }

def convert_from_color_segmentation(arr_3d):
    arr_2d = np.zeros((arr_3d.shape[0],arr_3d.shape[1]),dtype=np.uint8)
    
    for c,i in palette.items():
        m = np.all(arr_3d == np.array(c).reshape(1,1,3),axis=2)
        arr_2d[m]=i
    
    return arr_2d


# def get_index(cols_or_rows,step,_range):
#     index_temp1 = []
#     index_temp2 = []  ##hang or lie
#     index_temp3 = [] ##suiji
#     for i in range(0,cols_or_rows-step+1):
#         if(i%step == 0):  #保证覆盖全图
#             index_temp1.append(i)
#         elif((i-step/2)>=0 and (i-step/2)%step ==0):  ##交界处
#             index_temp2.append(i)
#         else:
#             index_temp3.append(i)

#     np.random.shuffle(index_temp3)   
#     return (index_temp1+index_temp2+index_temp3)[0:_range]
# #    return (index_temp1+index_temp2)

def get_index2(cols_or_rows,step,block_col_or_row,_range):
    index_temp1 = []
    index_temp2 = []
    for i in range(0,cols_or_rows-block_col_or_row+1,step):  #######step*2??
        index_temp1.append(i)
        
    for i in range(0,cols_or_rows-block_col_or_row):
        if(i%step != 0):
            index_temp2.append(i)
    
    np.random.shuffle(index_temp2)
    
    return (index_temp1+index_temp2)[0:_range]


padding_row=60
padding_col=64
step_row = 60
step_col = 64
block_row=240
block_col=256
range_row = 50  ##40
range_col = 60  ##50

print "start"

# f1=file("train.txt", "a+")
# f2=file("val.txt", "a+")
for n in range(2,52):
    count = 1
    
    srcPath = "data1/Padding/src-%d.jpg"%(n)
    labelPath = "data1/Padding/label-%d.png"%(n)
    
    img_src = Image.open(srcPath)
    img_src_RGB = img_src.convert("RGB")
    img_label = Image.open(labelPath)
    img_label_RGB = img_label.convert("RGB")
    
    arr_src = np.array(img_src_RGB)
    arr_label = np.array(img_label_RGB)

    ##将标签图片转换成label图
    labelTemp = convert_from_color_segmentation(arr_label)
    #Image.fromarray(labelTemp).save("data4_all/label_big/label-%d.png"%(n))
    #print arr_label.shape

    ###每一块的起始点位置索引
    index_rows = get_index2(arr_src.shape[0],step_row,block_row,range_row)
    index_cols = get_index2(arr_src.shape[1],step_col,block_col,range_col)
    

    for y in index_rows:
        for x in index_cols:
            #输出训练样本
            imgTemp = np.zeros((block_row,block_col,3),np.uint8)
            for i in range(0,block_row):
                for j in range(0,block_col):
                    imgTemp[i][j][0] = arr_src[y + i][x + j][0]
                    imgTemp[i][j][1] = arr_src[y + i][x + j][1]
                    imgTemp[i][j][2] = arr_src[y + i][x + j][2]
            outName1 = ""
            outName1 = "data2/train/%d_%d.jpg"%(n,count)
            Image.fromarray(imgTemp).save(outName1)
#                 #train.txt
#                 new_context = "%d"%(count) + '\n'
#                 f.write(new_context)

            #输出标签样本
            label_small_Temp = np.zeros((block_row,block_col),dtype=np.uint8)
            for i in range(0,block_row):
                for j in range(0,block_col):
                    label_small_Temp[i][j] = labelTemp[y + i][x + j]

            outName2 = ""
            outName2 = "data2/label/%d_%d.png"%(n,count)
            Image.fromarray(label_small_Temp).save(outName2)


#             #输出txt
#             if(count%10 != 0):
#                 new_context = "%d"%(count) + '\n'
#                 f1.write(new_context)  ###train.txt
#             else:
#                 new_context = "%d"%(count) + '\n'
#                 f2.write(new_context)  ###var.txt

            count+=1
                
    print ("%d done"%(n))
    print count

# f1.close()
# f2.close()
#print count
print "All done"
```
